```python
import whisper
from whisper.utils import write_vtt
```

```python
model = whisper.load_model('large')
```

```python
a5 = '/content/GB_FM_5a.mp3'
```

```python
a5r = model.transcribe(a5, language="Spanish")
```

```python
with open('5a.vtt', "w") as txt:
  write_vtt(a5r["segments"], file=txt)
```

```python
with open('5a.vtt', "r") as txt:
 print('\n'.join(txt.readlines()))
```

```python
from whisper.utils import write_txt

with open('5a.txt', "w") as txt:
  write_txt(a5r['segments'], file=txt)
```
